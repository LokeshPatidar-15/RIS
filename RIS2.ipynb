```python
import matplotlib.pyplot as plt
import numpy as np

# SNR values in dB
snr_db = np.arange(0, 21, 2)

# Simulated BER values (example for illustration)
ber_no_ris =     [1e-1, 9e-2, 7.5e-2, 6e-2, 5e-2, 4.2e-2, 3.5e-2, 3e-2, 2.5e-2, 2e-2, 1.5e-2]
ber_single_ris = [8e-2, 6e-2, 4.5e-2, 3.2e-2, 2.1e-2, 1.4e-2, 9e-3, 6e-3, 3.5e-3, 2e-3, 1e-3]
ber_dual_ris =   [5e-2, 3.5e-2, 2.2e-2, 1.5e-2, 1e-2, 6e-3, 3e-3, 1.8e-3, 1e-3, 6e-4, 2.5e-4]

# Plotting
plt.figure(figsize=(9, 6))
plt.semilogy(snr_db, ber_no_ris, 'r^-', label='No RIS')
plt.semilogy(snr_db, ber_single_ris, 'bo-', label='Single RIS')
plt.semilogy(snr_db, ber_dual_ris, 'gs--', label='Dual RIS')

plt.xlabel('SNR (dB)', fontsize=12)
plt.ylabel('Bit Error Rate (BER)', fontsize=12)
plt.title('BER vs SNR for No RIS, Single RIS, and Dual RIS', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Distance in meters
distance = np.arange(10, 120, 10)  # from 10m to 110m

# Simulated BER values for each configuration
ber_no_ris =     [1e-2, 2e-2, 3.5e-2, 5e-2, 6.5e-2, 8e-2, 9e-2, 1.1e-1, 1.3e-1, 1.5e-1, 1.7e-1]
ber_single_ris = [8e-3, 1.5e-2, 2.5e-2, 3.5e-2, 4.5e-2, 5.5e-2, 6.5e-2, 7.5e-2, 8.5e-2, 9.5e-2, 1.05e-1]
ber_dual_ris =   [5e-3, 9e-3, 1.4e-2, 2e-2, 2.7e-2, 3.3e-2, 3.9e-2, 4.5e-2, 5.2e-2, 6e-2, 6.8e-2]

# Plotting
plt.figure(figsize=(9, 6))
plt.semilogy(distance, ber_no_ris, 'r^-', label='No RIS')
plt.semilogy(distance, ber_single_ris, 'bo-', label='Single RIS')
plt.semilogy(distance, ber_dual_ris, 'gs--', label='Dual RIS')

plt.xlabel('Distance L (meters)', fontsize=12)
plt.ylabel('Bit Error Rate (BER)', fontsize=12)
plt.title('BER vs Distance (L) for No RIS, Single RIS, and Dual RIS', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Epochs
epochs = np.arange(1, 51)

# Simulated training loss (exponential decay with slight noise)
np.random.seed(0)
loss = 0.8 * np.exp(-0.07 * epochs) + np.random.normal(0, 0.01, len(epochs))

# Plot
plt.figure(figsize=(9, 6))
plt.plot(epochs, loss, 'b-', linewidth=2, label='Training Loss')

plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.title('Loss vs Epoch', fontsize=14)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
